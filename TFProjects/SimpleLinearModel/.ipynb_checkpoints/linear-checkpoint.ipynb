```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from sklearn.metrics import confusion_matrix
```

```python
from tensorflow.examples.tutorials.mnist import input_data
data = input_data.read_data_sets("data/MNIST/", one_hot=True)
```

##### The MNIST data-set has now been loaded and consists of 70,000 images and associated labels (i.e. classifications of the images.) The data-set is split into 3 mutually exclusive sub-sets. We only use the trainning and test-sets here.

```python
print("Size of:")
print("- Training-set:\t\t{}".format(len(data.train.labels)))
print("- Test-ste:\t\t{}".format(len(data.test.labels)))
print("- Validation-set:\t\t{}".format(len(data.validation.labels)))
```

##### One-Hot Encoding
The data-set has been loaded as so-called One-Hot encoding. This means the labels have been converted from a single number to a vector whose length equals the number of possible classes. All elements of the vector are zero except for the ith element which is one and means the class is i. For example, the One-Hot encoded labels for the first 5 images in the test-set are:

```python
data.test.labels[0:5, ]
```

```python
data.test.cls = np.array([label.argmax() for label in data.test.labels])
try_cls = np.argmax(data.test.labels, axis=1)
```

```python
print(data.test.cls[0:5])
print(try_cls[0:5])
```

##### Data dimensions:

```python
img_size = 28
img_size_flat = img_size**2
img_shape = (img_size, img_size)
num_classes = len(data.test.labels[0])
print(img_size, img_size_flat, img_shape, num_classes)
```

##### Plotting functions:
Function used to plot 9 images in a rXc grid, and writing the true and predicted classes below each image

```python
def plotImages(r, c, images, cls_true, cls_pred=None):
    assert len(images) == len(cls_true) == r*c
    assert r > 0 and c > 0
    
    # create subplots with r x c grid:
    fig, axes = plt.subplots(r, c)
    fig.subplots_adjust(hspace = 0.3, wspace = 0.3)
    
    print(axes.flat[0])
    
    for i, ax in enumerate(axes.flat):
        # Plot image.
        ax.imshow(images[i].reshape(img_shape), cmap='binary')
        
        # Shw true and predicted classes.
        if cls_pred is None:
            xlabel = "True: {0}".format(cls_true[i])
        else:
            xlabel = "True: {0}, Pred:{1}".format(cls_true[i], cls_pred[i])
        
        ax.set_xlabel(xlabel)
        
        # Remove ticks from the plot.
        ax.set_xticks([])
        ax.set_yticks([])
```

Plot a few images

```python
imgs = data.test.images[0:12]
cls_true = data.test.cls[0:12]
plotImages(3, 4, images = imgs, cls_true = cls_true)
```

##### Tensorflow and computational graphs
1. Placeholder variables used to change the input to the graph
2. Model variables that are going to be optimized so as to make the model perform better.
3. The model which is essentially just a mathematical function that calculates some output given the input in the placeholder variables and the model variables.
4. A cost measure that can be used to guide the optimization of the variables.
5. An optimization method which updates the variables of the model.

In addition, the Tensorflow graph may also cantain various debugging statements e.g. for logging data to the displayed using Tensorboard.

```python
'''Define Placeholder variables'''
x = tf.placeholder(tf.float32, [None, img_size_flat])
y_true = tf.placeholder(tf.float32, [None, num_classes])
y_true_cls = tf.placeholder(tf.int64, [None])
```

```python
'''Define variables to be optimized'''
weights = tf.Variable(tf.zeros([img_size_flat, num_classes]))
biases = tf.Variable(tf.zeros([num_classes]))
```

##### Model
 The simple mathematical model multiplies the images in the placeholder variable x with the weights and then adds the biases.

```python
'''The name logits is typical Tensorflow 
terminology, but other people may call the 
variable something else.'''
logits = tf.matmul(x, weights) + biases
```

```python
'''Using softmax to output the final results'''
y_pred = tf.nn.softmax(logits)
y_pred_cls = tf.argmax(y_pred, axis=1)
```

```python
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(
    logits = logits, labels = y_true)
cost = tf.reduce_mean(cross_entropy)
```

##### Optimization section:

```python
optimizer = tf.train.GradientDescentOptimizer(
    learning_rate = 0.5).minimize(cost)
'''Define Performance measure'''
correct_prediction = tf.equal(y_pred_cls, y_true_cls)
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
'''Up to now, there has no calculation been done.'''
```

###### The following it the tensorflow run section 

```python
'''Create Tensorflow session'''
session = tf.Session()
'''Initialize variable'''
#session.run(tf.initialize_all_variables())
session.run(tf.global_variables_initializer())
```

##### Define a help function to better do optimization

```python
batch_size = 100
def optimize(num_iterations):
    for i in range(num_iterations):
        '''Using stochastic optimization'''
        x_batch, y_true_batch = data.train.next_batch(
            batch_size)
        '''These batches are input for this iteration'''
        feed_dict_train = {
            x: x_batch,
            y_true: y_true_batch,
        }
        '''Run the optimizer using this batch of training data'''
        session.run(optimizer, feed_dict=feed_dict_train)
```

```python
'''Check performances'''
feed_dict_test = {
    x: data.test.images,
    y_true: data.test.labels,
    y_true_cls: data.test.cls
}
def printAccuracy():
    acc = session.run(accuracy, feed_dict = feed_dict_test)
    # Print the accuracy.
    print("Accuracy on test-set: {0:.1%}".format(acc))
```

```python
'''Define confusion matrix'''
def printConfusionMatrix():
    # get true classification
    cls_true = data.test.cls
    
    # get the predictions
    cls_pred = session.run(y_pred_cls, feed_dict=feed_dict_test)
    
    # Get the confusion matrix using sklearn.
    cm = confusion_matrix(y_true = cls_true, y_pred = cls_pred)
    
    # Print the confusion matrix as text
    print(cm)
    
    # Plot the confusion matrix as an image.
    plt.imshow(cm, interpolation = 'nearest', cmap = plt.cm.Blues)
    
    #Make various adjustments to the plot.
    plt.tight_layout()
    plt.colorbar()
    tick_marks = np.arange(num_classes)
    plt.xticks(tick_marks, range(num_classes))
    plt.yticks(tick_marks, range(num_classes))
    plt.xlabel('Predicted')
    plt.ylabel('True')
```

```python
'''Plot example errors'''
def plotExampleErrors():
    # Use TensorFlow to get a list of boolean values
    # whether each test-image has been correctly classified,
    # and a list for the predicted class of each image.
    correct, cls_pred = session.run([correct_prediction, y_pred_cls],
                                    feed_dict=feed_dict_test)

    # Negate the boolean array.
    incorrect = (correct == False)
    
    # Get the images from the test-set that have been
    # incorrectly classified.
    images = data.test.images[incorrect]
    
    # Get the predicted classes for those images.
    cls_pred = cls_pred[incorrect]

    # Get the true classes for those images.
    cls_true = data.test.cls[incorrect]
    
    # Plot the first 9 images.
    plotImages(3, 3, images=images[0:9],
                cls_true=cls_true[0:9],
                cls_pred=cls_pred[0:9])
```

```python
'''Plot Weights'''
def plotWeights():
    # Get the values for the weights from the TensorFlow variable.
    w = session.run(weights)
    
    # Get the lowest and highest values for the weights.
    # This is used to correct the colour intensity across
    # the images so they can be compared with each other.
    w_min = np.min(w)
    w_max = np.max(w)

    # Create figure with 3x4 sub-plots,
    # where the last 2 sub-plots are unused.
    fig, axes = plt.subplots(3, 3)
    fig.subplots_adjust(hspace=0.3, wspace=0.3)

    for i, ax in enumerate(axes.flat):
        # Only use the weights for the first 10 sub-plots.
        if i<9:
            # Get the weights for the i'th digit and reshape it.
            # Note that w.shape == (img_size_flat, 10)
            image = w[:, i].reshape(img_shape)

            # Set the label for the sub-plot.
            ax.set_xlabel("Weights: {0}".format(i))

            # Plot the image.
            ax.imshow(image, vmin=w_min, vmax=w_max, cmap='seismic')

        # Remove ticks from each sub-plot.
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
printAccuracy()
```

```python
plotExampleErrors()
```

```python
optimize(num_iterations=1)
printAccuracy()
plotExampleErrors()
plotWeights()
```

```python
optimize(num_iterations=9)
printAccuracy()
plotExampleErrors()
plotWeights()
```
